```python
import wfdb
import matplotlib.pyplot as plt
import numpy as np
from scipy.signal import butter, filtfilt
```

```python
# Set the path to the MIT-BIH Arrhythmia Database
mitdb_path = '/content/drive/MyDrive/mit-bih-arrhythmia-database-1.0.0/mit-bih-arrhythmia-database-1.0.0'
filenumb = 100

# Read the signal and annotation data 
record = wfdb.rdrecord(mitdb_path +  '/' + str(filenumb), sampto=3600, channels=[0, 1])

# Read the annotation data
annotations = wfdb.rdann(mitdb_path + '/' +str(filenumb), 'atr', sampto=3600)

# Print the signal data
print(record.p_signal)

# Print the annotation data
print(annotations.symbol)
print(len(t))
print(N)
```

```python
N = len(record.p_signal) #number of data point
sr = 360 #sample rate
t_max = N/sr #maximum time
dt = 1/sr #timestep
t = np.arange(0,t_max,dt)
```

```python
# Plot the ECG graph
plt.plot(t,record.p_signal[:,0])
plt.rcParams['figure.figsize'] = [20,5]
plt.xlabel('Time (s)')
plt.ylabel('Voltage (mV)')
plt.title('File Number: ' + str(filenumb))
plt.show()
```

```python
# Get the ECG signal data
ecg_signal = record.p_signal[:,0]

# Define the filter parameters
fs = record.fs
lowcut = 0.5
highcut = 40

# Apply a Butterworth filter to the ECG signal
nyquist = 0.5 * fs
low = lowcut / nyquist
high = highcut / nyquist
order = 4
b, a = butter(order, [low, high], btype='band')
filtered_ecg_signal = filtfilt(b, a, ecg_signal)

# Apply FFT to the filtered ECG signal
fft_ecg_signal = np.fft.fft(filtered_ecg_signal)
freq = np.fft.fftfreq(N,dt)

PSD = np.abs(fft_ecg_signal)**2

# Plot the power spectrum
plt.plot(freq,PSD)
plt.xlabel('Frequency (Hz)')
plt.title('Power Spectral Density File Number: ' + str(filenumb))
plt.ylabel('PSD (mV^2)')
plt.xlim(0, 20)
plt.show()
```

```python
#Gaussian Function
def gaussian (A, mu, sigma, x):
  M = (x-mu)**2/(2*sigma**2)
  return A*np.exp(-M)
```

```python
#Parameter
A = 1
mu = 7.5
sigma = 3

f_gaussian = gaussian(A, mu, sigma, freq)

plt.plot(freq,f_gaussian,color='k',linewidth="1.5")
plt.xlim(0,20)
plt.legend()
plt.show()
```

```python
#filtering with gaussian
fhat_window = fft_ecg_signal*f_gaussian

ifhat = np.fft.ifft(fhat_window)

plt.plot(t,ifhat,color='k',linewidth="1.5")
plt.xlim(0,10)
plt.legend()
plt.show()
```

```python
from scipy.signal import find_peaks

```

```python
peaks, _ = find_peaks(ifhat, height=0.05)
t_peak = peaks/sr

plt.plot(t,ifhat)
plt.plot(t_peak, ifhat[peaks], "or")
plt.xlim(0,10)
plt.show()
```
